# Advanced Feature Engineering

## Objectives

Enhance the dataset by creating additional technical indicators that may improve machine learning model performance.

```python
import numpy as np
import pandas as pd
```

```python
stock = pd.read_csv(
    "../data/processed/reliance_features.csv",
    parse_dates=["Date"]
)

stock.set_index("Date", inplace=True)

stock.head()
```

```python
delta = stock["Close"].diff()

gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()

rs = avg_gain / avg_loss

stock["RSI"] = 100 - (100 / (1 + rs))
```

```python
ema12 = stock["Close"].ewm(span=12, adjust=False).mean()
ema26 = stock["Close"].ewm(span=26, adjust=False).mean()

stock["MACD"] = ema12 - ema26

stock["MACD_Signal"] = stock["MACD"].ewm(
    span=9,
    adjust=False
).mean()
```

```python
rolling_mean = stock["Close"].rolling(20).mean()
rolling_std = stock["Close"].rolling(20).std()

stock["BB_Upper"] = rolling_mean + (2 * rolling_std)
stock["BB_Lower"] = rolling_mean - (2 * rolling_std)

stock["BB_Width"] = stock["BB_Upper"] - stock["BB_Lower"]
```

```python
stock["Momentum10"] = (
    stock["Close"] - stock["Close"].shift(10)
)
```

```python
stock = stock.dropna()
```

```python
stock.columns.tolist()
```

```python
stock.to_csv("../data/processed/reliance_advanced_features.csv")
```
