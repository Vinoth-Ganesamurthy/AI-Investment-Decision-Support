## Notebook 01 – Stock Data Download

This notebook downloads historical stock price data for Reliance Industries Ltd. using the Yahoo Finance API (`yfinance`) and stores it locally for further analysis.

```python
import pandas as pd
import yfinance as yf
```

```python
ticker = "RELIANCE.NS"

stock = yf.download(
    ticker,
    start="2015-01-01",
    end="2024-12-31",
    auto_adjust=False
)
```

```python
stock.head()
```

```python
stock.info()
```

```python
stock.to_csv("../data/raw/reliance.csv")
```

```python
saved_stock = pd.read_csv("../data/raw/reliance.csv")

saved_stock.head()
```
