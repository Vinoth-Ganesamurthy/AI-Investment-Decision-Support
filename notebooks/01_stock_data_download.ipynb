## Notebook 01 - Stock Data Download

This notebook downloads historical stock price data for Reliance Industries and saves it for further analysis.

```python
import pandas as pd
import yfinance as yf
```

```python
ticker = "RELIANCE.NS"

stock = yf.download(
    ticker,
    start="2015-01-01",
    end="2024-12-31",
    auto_adjust=False
)
```

```python
stock.columns = stock.columns.get_level_values(0)

stock.columns
```

```python
stock = stock.reset_index()

stock.head()
```

```python
print(stock.columns.tolist())
```

```python
stock.to_csv("../data/raw/reliance.csv", index=False)

print("Reliance stock data saved successfully!")
```
