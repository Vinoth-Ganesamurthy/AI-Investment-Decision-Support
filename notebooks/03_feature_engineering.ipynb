# Feature Engineering

## Objectives

Create technical indicators that can be used as features for machine learning models.

```python
import numpy as np
import pandas as pd
```

```python
stock = pd.read_csv(
    "../data/raw/reliance.csv",
    parse_dates=["Date"]
)

stock.set_index("Date", inplace=True)

stock.head()
```

```python
stock["Daily_Return"] = stock["Close"].pct_change()
```

```python
stock["Log_Return"] = np.log(
    stock["Close"] / stock["Close"].shift(1)
)
```

```python
stock["Trading_Range"] = stock["High"] - stock["Low"]
```

```python
stock["Price_Change"] = stock["Close"] - stock["Open"]
```

```python
stock["Volatility20"] = (
    stock["Daily_Return"]
    .rolling(window=20)
    .std()
)
```

```python
stock["MA20"] = stock["Close"].rolling(20).mean()

stock["MA50"] = stock["Close"].rolling(50).mean()

stock["MA200"] = stock["Close"].rolling(200).mean()
```

```python
stock["EMA20"] = stock["Close"].ewm(
    span=20,
    adjust=False
).mean()
```

```python
stock["Target"] = (
    stock["Close"].shift(-1) > stock["Close"]
).astype(int)
```

```python
stock = stock.dropna()
```

```python
stock.to_csv("../data/processed/reliance_features.csv")
```
